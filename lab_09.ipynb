```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def load_image():
    blank_image = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_image,text='FIGHT',org=(50,300),fontFace=font,fontScale = 5,color=(255,255,255),thickness=15,lineType=cv2.LINE_AA)
    return blank_image
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img=load_image()
```

```python
display_img(img)
```

```python
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
erosion = cv2.erode(img,kernel,iterations = 2)
```

```python
display_img(erosion)
```

```python
dilution=cv2.dilate(img,kernel,iterations=2)
```

```python
display_img(dilution)
```
